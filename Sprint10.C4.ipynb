```python
average_seating_capacity = df['Seating_Capacity'].mean()

# Count occurrences of central locking (assuming 1 for yes, 0 for no)
central_locking_count = df['Central_Locking'].value_counts()

# Count occurrences of child safety locks (assuming 1 for yes, 0 for no)
child_safety_locks_count = df['Child_Safety_Locks'].value_counts()

# Display results
display(f"Average Seating Capacity: {average_seating_capacity:.2f}")

# Display counts in a DataFrame format for better visualization
display(central_locking_count)
display(child_safety_locks_count)
```

```python
#Task 3: Alert Systems Analysis

#Investigate the presence of high-speed alerts, seat belt reminders, and door-ajar warnings to assess the integration of safety and convenience features with respect to each feature mentioned above.
```

```python
high_speed_alert_count = df['High_Speed_Alert_System'].value_counts()
seat_belt_reminder_count = df['Fasten_Seat_Belt_Warning'].value_counts()
door_ajar_warning_count = df['Door_Ajar_Warning'].value_counts()

# Display results
display("High-Speed Alert System Count:")
display(high_speed_alert_count)

display("Seat Belt Reminder Count:")
display(seat_belt_reminder_count)

display("Door Ajar Warning Count:")
display(door_ajar_warning_count)
```

```python
#Task 4: Dimensional Analysis

#Analyze the dimensions of cars (height, length, width) to understand size preferences and market demands, aiding in product planning.

#Also, present make-wise average summaries for all dimensions of cars analyzed above
```

```python
# Calculate make-wise average summaries for dimensions
average_dimensions = df.groupby('Make')[['Height_mm', 'Length_mm', 'Width_mm']].mean().reset_index()

# Rename the columns for better clarity
average_dimensions.columns = ['Make', 'Average Height (mm)', 'Average Length (mm)', 'Average Width (mm)']

# Display the results DataFrame
display(average_dimensions)
```
