```python
#Task 12

#Compare the performance of cars based on factors such as displacement, cylinders, and valves per cylinder, providing valuable insights for marketing and product development.

#Hint: Calculate and display the mean, median, standard deviation, and coefficient of variation for each manufacturer.

#For the "Displacement" column, display the distribution for each manufacturer using box plots.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset (adjust the path accordingly)
df = pd.read_excel(r'C:\Users\91994\Downloads\DS1_C4_S9_Car_Data.xlsx')

# Calculating statistics
stats = df.groupby('Make').agg(
    Displacement_mean=('Displacement', 'mean'),
    Displacement_median=('Displacement', 'median'),
    Displacement_std=('Displacement', 'std'),
    Cylinders_mean=('Cylinders', 'mean'),
    Cylinders_median=('Cylinders', 'median'),
    Cylinders_std=('Cylinders', 'std'),
    Valves_Per_Cylinder_mean=('Valves_Per_Cylinder', 'mean'),
    Valves_Per_Cylinder_median=('Valves_Per_Cylinder', 'median'),
    Valves_Per_Cylinder_std=('Valves_Per_Cylinder', 'std'),
)

# Calculate Coefficient of Variation
stats['Displacement_cv'] = stats['Displacement_std'] / stats['Displacement_mean']
stats['Cylinders_cv'] = stats['Cylinders_std'] / stats['Cylinders_mean']
stats['Valves_Per_Cylinder_cv'] = stats['Valves_Per_Cylinder_std'] / stats['Valves_Per_Cylinder_mean']

# Display statistics
stats.reset_index()
```

```python
# Create a box plot for "Displacement" using Seaborn
plt.figure(figsize=(12, 6))
sns.boxplot(x='Make', y='Displacement', data=df)
plt.title('Displacement Distribution by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Displacement (cc)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
